```python
import pandas as pd
import glob
import os

# Path to the folder containing your CSV files
folder_path = "final_processed"  # Replace with the actual folder path
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# List to store DataFrames
dfs = []

# Loop through each file and add 'subject' column
for file in csv_files:
    # Read the CSV file without any headers
    df = pd.read_csv(file, header=None)
    
    # Add a 'subject' column with the file name (or any other unique identifier)
    df['subject'] = os.path.basename(file).split('.')[0]
    
    # Append to the list of DataFrames
    dfs.append(df)

# Concatenate all DataFrames without headers
combined_df = pd.concat(dfs, ignore_index=True)

# Assign column names after concatenation
# Assuming 1000 columns in original data, add 1 for 'subject'
combined_df.columns = [f"roi_{j}" for j in range(combined_df.shape[1] - 1)] + ['subject']

# Save to a single CSV file without an extra header
combined_df.to_csv("combined_file_with_subjects.csv", index=False)
print("All files have been combined into combined_file_with_subjects.csv")
```
