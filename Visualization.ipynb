```python
import nibabel as nib
import numpy as np
from nilearn import plotting

# Load the fMRI image (4D image)
fmri_image = nib.load(root_path)  # Replace with your actual fMRI image path
frame_data = fmri_image.get_fdata()  # Load the data

# Ensure the data shape is correct
data_shape = frame_data.shape
if len(data_shape) != 4:
    raise ValueError("Expected a 4D fMRI image. Please check the input data.")

# Select a single frame (e.g., the middle slice in the time dimension)
frame_index = data_shape[3] // 2  # Access the time dimension correctly
single_frame = frame_data[:, :, :, frame_index]  # Select the 3D slice for the middle time point

# Create a NIfTI image for the single frame
single_frame_img = nib.Nifti1Image(single_frame, affine=fmri_image.affine)

# Use Nilearn to plot the 3D view
plotting.view_img(single_frame_img, threshold='auto', title='3D Visualization of fMRI Data (Single Frame)')
```
